
# Oracle Database CPU Exploratory Data Analysis Using Kernel Density Estimation #

In this example we are trying to do kernel density approach to database time modelling. If you are into tuning database systems, one of the most important statistics we care about is time. We can check for time statistics in Time Model Statistics in AWR Reports. It looks like the snapshot below.

```python
Image(filename='images/kde_time_model.png')
```

DB CPU is the oracle server CPU consumption. DB CPU in AWR is extracted from the views v\$sess_time_model and v\$sys_time_model.

DB CPU is the amount of CPU consumed on user level calls. This doesn't amounts to the CPU time spent on the background processes and is restricted to only foregrounds.

DB time is the total time spent in the database by foreground sessions including CPU Time, IO time and wait time.
