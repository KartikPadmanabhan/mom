```python
df=pd.read_csv('data/TMP_AWR_WAIT_CLASS_TWT_AGG.csv', sep='|',parse_dates=True)
df.head()
```

```python
df=df.pivot_table(index='start_time', columns='stat_name', values='stat_waits')
df.head()
```

```python
df.info()
```

Replacing all the null values by its mean

```python
columns=df.columns
for column in columns:
    mean_column=np.mean(df[column])
    df[column]=df[column].fillna(mean_column)
```

```python
dist_names = ['gamma', 'lognorm', 'alpha' ,'anglit','arcsine','bradford','burr', 'cauchy']
```

```python
import scipy
size = 30000
#dist_names = ['gamma', 'lognorm', 'alpha' ,'anglit']
dist_names = ['alpha']
x = np.linspace(df['Application'].min(), df['Application'].max())
y=df['Application'].values
h = plt.hist(y,  bins=50, color='b', alpha=0.5)

for dist_name in dist_names:
    dist = getattr(scipy.stats, dist_name)
    param = dist.fit(y)
    pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1]) * size
    plt.plot(pdf_fitted, label=dist_name)
plt.legend(loc='upper right')
plt.show()
```

```python
size = 30000
x = scipy.arange(size)
y = scipy.int_(scipy.round_(scipy.stats.vonmises.rvs(5,size=size)*47))

h = plt.hist(y, bins=range(48), color='w')

dist_names = ['gamma', 'beta', 'rayleigh', 'norm', 'pareto']

for dist_name in dist_names:
    dist = getattr(scipy.stats, dist_name)
    param = dist.fit(y)
    pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1]) * size
    plt.plot(pdf_fitted, label=dist_name)
    plt.xlim(0,47)
plt.legend(loc='upper right')
```
